```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("../data/sales.csv", parse_dates=["order_date"])
df.head()
```

```python
#features business
df['revenue'] = df['quantity'] * df['unit_price']
df['revenue_after_discount'] = df['revenue'] * (1 - df['discount'])
```

### KPI Business

```python
total_revenue = df["revenue_after_discount"].sum()

monthly_revenue = (
    df.groupby(df["order_date"].dt.to_period("M"))["revenue_after_discount"]
    .sum()
)

average_order_values = (
    df.groupby("order_id")["revenue_after_discount"]
    .sum()
    .mean()
)

discount_impact =(df["revenue"]-df["revenue_after_discount"]).sum()

total_revenue, monthly_revenue, average_order_values,discount_impact
```

###### total_revenue → CA net réel
###### monthly_revenue → tendance mensuelle
###### average_order_values (AOV) → panier moyen
###### discount_impact → perte de CA due aux remises

### Graphe Business:CA mensuel

```python
#creation du graphique
plt.figure()
monthly_revenue.plot(marker="o")
plt.title("Monthly revenue trend")
plt.xlabel("Month")
plt.ylabel("Revenue (euro)")
plt.tight_layout()

#sauvegarde 
plt.savefig("../reports/monthly_revenue_business.png")
plt.show()
```

### Impact des Remises

```python
df["discounr_level"] = pd.cut(
    df["discount"], 
    bins=[-0.01, 0.05, 0.15, 1], 
    labels=[" low (<=5%)", "Medium (5-15%)", "Hight (>15%)"]
)

df.groupby("discounr_level")["revenue_after_discount"].sum()
```

### Conclusion Business 

```python
## Business insights

### Insight 1 – Revenue concentration
Le chiffre d’affaires est principalement généré par la catégorie Electronics, ce qui indique une forte dépendance à cette catégorie.

### Insight 2 – Discount impact
Les remises élevées (>15%) réduisent significativement le revenu net sans générer une augmentation proportionnelle des ventes.

### Insight 3 – Monthly trend
Le chiffre d’affaires mensuel montre une évolution stable, suggérant l’absence de forte saisonnalité sur la période analysée.

### Business recommendation
- Réduire les remises élevées
- Renforcer les ventes sur les catégories secondaires
- Développer les villes à fort potentiel mais faible volume
```
